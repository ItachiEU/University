```python
from collections import defaultdict

sentences = list()

tests = list()

with open('../../NLP_Resources/sentences_for_task1.txt') as f:
    for line in f:
        sentences.append(line)

with open('../../NLP_Resources/test_for_task1.txt') as f:
    for line in f:
        tests.append(tuple(line.split()))
```

```python
correctWords = set()

def treshold(length):
    if length <= 7:
        return (length - 6.2)**6 + 600/length
    return 150 / length

def findWord(sentence):

    bestLength = 1
    
    for length in range(2, min(len(sentence) + 1, 15)):
        currentWord = sentence[:length]
        temp = 0
        for line in sentences:
            x = line.find(currentWord)
            while x != -1:

                temp += 1
                line = line[x+1:]

                x = line.find(currentWord)

        if temp > treshold(length):
            bestLength = length

    return sentence[:bestLength]                



def fillCorrectWords():
    counter = 0
    for sentence in sentences:
        print(counter)
        counter += 1
        currentIndex = 0

        while currentIndex < len(sentence):
            wordFound = findWord(sentence[currentIndex:])
            correctWords.add(wordFound)
            currentIndex += len(wordFound)
        
```

```python
import copy

sufiksySorted = defaultdict(int)
prefiksySorted = defaultdict(int)

def insertSpaces(sentences, k):
    corp = copy.deepcopy(sentences)

    for i in range(len(corp)):
        curr = corp[i]
        curr = curr.strip()
        pref = curr[:k+1]
        suf = curr[-k-1:]
        sufiksySorted[suf] +=1
        prefiksySorted[pref] += 1
    
    prefiksy = list()
    sufiksy = list()

    for k in sufiksySorted.keys():
        sufiksy.append((sufiksySorted[k], k))
    
    for k in prefiksySorted.keys():
        prefiksy.append((prefiksySorted[k], k))

    prefiksy = sorted(prefiksy, reverse=True)
    sufiksy = sorted(sufiksy, reverse=True)

    for i in range(max(len(prefiksy), len(sufiksy))):
        if i < len(prefiksy):
            pref = prefiksy[i][1]
            for k in range(len(corp)):
                corp[k] = corp[k].replace(pref, " " + pref)
        if i < len(sufiksy):
            suf = sufiksy[i][1]
            for k in range(len(corp)):
                corp[k] = corp[k].replace(suf, suf+" ")
        if i % 100 == 0:
            print(i, '/', max(len(prefiksy), len(sufiksy)))

    # print("prefiksy: ", len(prefiksy), "sufiksy: ", len(sufiksy))
    # counter = 0
    # for pref in prefiksy:
    #     pref = pref[1]
    #     counter += 1
    #     if counter % 100 == 0:
    #         print(counter, '/', len(prefiksy))
    #     for k in range(len(corp)):
    #         corp[k] = corp[k].replace(pref, " "+pref)
    # print("prefiksy - done")
    # counter = 0
    # for suf in sufiksy:
    #     suf = suf[1]
    #     counter += 1
    #     if counter % 100 == 0:
    #         print(counter, '/', len(sufiksy))
    #     for k in range(len(corp)):
    #         corp[k] = corp[k].replace(suf, suf+" ")
    # print("sufiksy - done")
    return corp       
    
new_corp4_sorted = insertSpaces(sentences, 4) 
```

```python
def evalCorp(word, corp):
    sum = 0
    for line in corp:
        sum += line.count(word)
        sum += 15 * line.count(" " + word)
        sum += 100 * line.count(" " + word + " ")

    return sum

result = 0

counter = 0
for test in tests[:10000]:
    if counter % 100 ==0:
        print(counter / 10000)
    counter += 1
    w1, w2 = test
    w1 = w1[:6]
    w2 = w2[:6]

    score_w1 = evalCorp(w1, new_corp4_sorted)
    score_w2 = evalCorp(w2, new_corp4_sorted)

    #print(w1, w2, score_w1, score_w2)

    if score_w1 > score_w2:
        result += 1
    elif score_w1 == score_w2:
        result += 0.5
    
print(result / 10000)
```

```python
def evalCorp(word, corp):
    word1 = word[:6]
    word2 = word[-6:]
    #print(word1, word2)
    sum = 0
    for line in corp:
        sum += line.count(word2)
        sum += line.count(word1)
        sum += 15 * line.count(" " + word1)
        sum += 15 * line.count(word2 + " ")
        sum += 100 * line.count(word)
        sum += 1000 * line.count(" " + word + " ")

    return sum
```
