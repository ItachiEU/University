## Zadanie 11


```python
y = [1.0, 2.0, 4.0, 4.0, 5.0, 7.0, 8.0, 9.0]
x = [1.0, 3.0, 4.0, 6.0, 8.0, 9.0, 11.0, 14.0]

n = len(x)

sum1 = sum(x[i] * y[i] for i in range(n))
sum2 = sum(x[i] ** 2 for i in range(n))

avg_x = sum(x)/n
avg_y = sum(y)/n

a  = (sum1 - n*avg_x*avg_y) / (sum2 - n*(avg_x**2))
b = avg_y - a * avg_x
print(f'Prosta regresji to:\nTemp = {a} * x + {b}')
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = x
reg = [a*x[i] + b for i in range(n)]
ax.plot(x, reg);
plt.plot(x, y, 'o')
```

```python
import numpy as np
v = [54.3, 61.8, 72.4, 88.7, 118.6, 194.0]
p = [61.2, 49.5, 37.6, 28.4, 19.2, 10.1]

v_log = [np.log(x) for x in v]
p_log = [np.log(x) for x in p]

n = len(v)

avg_v = sum(v_log)/n
avg_p = sum(p_log)/n

a  = sum([(vi - avg_v)*(pi - avg_p) for vi, pi in zip(v_log, p_log)]) / sum([(vi - avg_v)**2 for vi in v_log])

k = -a

b = avg_p - a*avg_v

c = np.exp(b)
print(b)
print(c)

# P dla v = 100 
P = c/100**k
print(P)
```
