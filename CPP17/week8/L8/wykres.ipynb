```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = []

with open("data.txt") as f:
    for line in f:
        real, imag = line.split(';')
        data.append((float(real), float(imag)))
```

```python
data = np.array(data)
```

```python

plt.figure(figsize=(15,10))
plt.plot(data[:, 0], data[:, 1], c='black')
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.arange(-30, 30, 0.01), data[:, 0])
plt.plot(np.arange(-30, 30, 0.01), data[:, 1])
plt.xlim([-30, 30])
# niebieski - real, pomarańczowy - imaginare
```
